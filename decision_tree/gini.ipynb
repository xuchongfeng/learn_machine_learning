```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = range(0, 1)
y = x
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 1, 10)
plt.plot(x, map(lambda x: x if x < 0.5 else 1 - x, x), label="linear")
```

```python
x = np.arange(0, 1, 0.0001)
plt.plot(x, np.where(x <= 0.5, x, 1-x), label="")
plt.plot(x, 2 * x * (1 - x), label="")
plt.plot(x, - 1/2 * (x * np.log2(x) + (1-x) * np.log2(1-x)), label="")
plt.legend((u"classification error rate", u"gini rate", u"half entropy"))
plt.show()
```
